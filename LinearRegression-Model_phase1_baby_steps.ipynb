```python
#library import
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

# Using panda reading csv file sstored in same file and stored in data variable
data = pd.read_csv("Salary_Data.csv")
print(data)
```

```python
# reading YearofExperince as Independent value or x value
Independent=data[["YearsExperience"]]
# reading Salary as Dependent value or Y value
dependent=data[["Salary"]]

```

```python
#using Matplot lib ploting both x and y value in from Independent and dependent value
# using plt scattering x and y using Independent and dependent
plt.scatter(Independent,dependent)
# Setting Xaxis as yearofExperience
plt.xlabel("yearofExperience",fontsize=20)
# Setting Yaxis as Salary
plt.ylabel("Salary",fontsize=20)
#plotting all x and y values
plt.show()
```

```python
# using sklearn library selection model to train, test and split
from sklearn.model_selection import train_test_split
#using trin_test_split funtion splitting train data and test data of test size 1/3 rd of each
x_train, x_test, y_train, y_test = train_test_split(Independent, dependent,test_size=1/3,random_state=0)
```

```python
print(y_train)
```

```python
#importing Linea Regression algaritham from linear model
from sklearn.linear_model import LinearRegression
#Get Linearregression method
regression = LinearRegression()
#fit the regression on training sets
regression.fit(x_train, y_train)
```

```python
#slope or Weight of regression or weight
weight = regression.coef_
print("Weight of the model={}".format(weight))
# bais or c value of regression y = 9345 * X + 26816
bais = regression.intercept_
print("Intercept of the model={}".format(bais))
```

```python
#prediction of first test data on regression fit of x_train, y_train
y_pred=regression.predict(x_test)
y_pred
```

```python
#from this library get r2_score from metrics of sklearn
from sklearn.metrics import r2_score
#using r2_score method get Test and prection datasets
r_score=r2_score(y_test,y_pred)
```

```python
r_score
```

```python
#get pickle library to dump save regression model in file of.sav
import pickle
filename='finalized_model.sav'
pickle.dump(regression, open(filename, 'wb'))
```

```python
#from pickle load regression model and predict for given value
loaded_model = pickle.load(open('finalized_model.sav','rb'))
result=loaded_model.predict([[15]])
print(result)
```
